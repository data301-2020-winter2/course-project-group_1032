```python
sns.set_theme(style="ticks",font_scale=1.2)
df = pf.load_and_process("../../Dataset.csv")
```

We can look at the top 8 population groups to see how crime rate is affected.

```python
plot = sns.countplot(x="Population_Description", data=df, palette='bright', order=df['Population_Description'].value_counts().index[:8])
plot = plot.set_xticklabels(plot.get_xticklabels(), rotation=40, ha="right")
```

Very interestingly, there are as many thefts commited in cities with 10,000-24,999 as there are in cities from 500,000-999,999 and counties over 100,000. This might not say too much about the actual rate of crime, however, because there could be many more cities with lower population that just add up to the same amount of thefts.
